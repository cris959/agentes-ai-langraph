```python
import os
from google.colab import userdata
from google import genai

try:
    # Levantamos tu clave de los secretos de Colab
    api_key = userdata.get("GEMINI_KEY")

    # Inicializamos el cliente oficial de Google
    client = genai.Client(api_key=api_key)
    print("✅ Cliente de Google GenAI inicializado correctamente.")
except Exception as e:
    print(f"❌ Error al inicializar: {e}")
```

```python
def consultar_gemini(tema: str) -> str:
    """
    Función didáctica para enviar un prompt estructurado a Gemini 2.5 Flash
    """
    # Armamos el prompt dinámico con el tema que ingrese el usuario
    prompt = f"Explica de manera clara y actualizada cuáles son los impactos de la inteligencia artificial en el área de {tema}."

    try:
        # Invocamos al modelo que descubrimos que tenés activo y con cuota libre
        response = client.models.generate_content(
            model='gemini-2.5-flash',
            contents=prompt,
        )
        return response.text
    except Exception as e:
        return f"❌ Error en la petición: {e}"
```

```python
import os
from google.colab import userdata
from langchain_core.tools import tool
from langchain_tavily import TavilySearch

# Inyectamos la API Key en el entorno
os.environ["TAVILY_API_KEY"] = userdata.get("TAVILY_API_KEY")

try:
    # Usamos la clase moderna recomendada por LangChain
    web_search_tool = TavilySearch(max_results=3)
    print("✅ Herramientas de Tavily configuradas con éxito (Versión Moderna 🚀).")
except Exception as e:
    print(f"❌ Error al configurar Tavily: {e}")
```

```python
print("🌐 Consultando a Tavily en tiempo real...")

try:
    # Definimos la búsqueda
    query_busqueda = "resultado Estados Unidos vs Paraguay mundial"

    # Invocamos la herramienta
    resultados_raw = web_search_tool.invoke(query_busqueda)

    print("\n📄 Contenido crudo devuelto por Tavily:")
    print("-" * 60)
    # Imprimimos directamente la variable para ver qué tipo de datos es
    print(resultados_raw)
    print("-" * 60)
    print(f"💡 Tipo de dato: {type(resultados_raw)}")

except Exception as e:
    print(f"❌ Error al buscar en internet: {e}")
```

```python
import os
from google.colab import userdata
from langchain_google_genai import ChatGoogleGenerativeAI
from langgraph.prebuilt import create_react_agent

# 1. Aseguramos las credenciales
os.environ["GOOGLE_API_KEY"] = userdata.get("GEMINI_KEY")
os.environ["TAVILY_API_KEY"] = userdata.get("TAVILY_API_KEY")

try:
    # 2. Inicializamos Gemini 2.5 Flash
    model = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)

    # 3. Cargamos la herramienta
    tools = [web_search_tool]

    # 4. Creamos el agente de la forma más básica y compatible posible
    agent_executor = create_react_agent(model, tools)

    print("🤖 ¡Agente Autónomo ReAct de LangGraph inicializado con éxito y blindado contra errores de argumentos!")

except Exception as e:
    print(f"❌ Error al configurar el agente: {e}")
```

```python
# Hacemos una pregunta que requiera información actual
consulta = "¿Cuando juega Argentina en el mundial?"

print(f"👤 Usuario: {consulta}\n")
print("⏳ El agente está pensando y decidiendo qué herramientas usar...")

# Ejecutamos el agente pasándole la consulta en el estado
inputs = {"messages": [("user", consulta)]}
response = agent_executor.invoke(inputs)

# LangGraph nos devuelve todo el historial de mensajes (los pensamientos de la IA, el uso de la herramienta y la respuesta)
# Vamos a extraer directamente el último mensaje, que es la respuesta final para el usuario
respuesta_final = response["messages"][-1].content

print("\n🤖 Respuesta Final del Agente:")
print("-" * 60)
print(respuesta_final)
print("-" * 60)
```
