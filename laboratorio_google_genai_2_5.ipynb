```python
import os
from google.colab import userdata
from langchain_google_genai import ChatGoogleGenerativeAI
from langgraph.prebuilt import create_react_agent

# 1. Aseguramos las credenciales
os.environ["GOOGLE_API_KEY"] = userdata.get("GEMINI_KEY")
os.environ["TAVILY_API_KEY"] = userdata.get("TAVILY_API_KEY")

try:
    # 2. Inicializamos Gemini 2.5 Flash
    model = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)

    # 3. Cargamos la herramienta
    tools = [web_search_tool]

    # 4. Creamos el agente de la forma más básica y compatible posible
    agent_executor = create_react_agent(model, tools)

    print("🤖 ¡Agente Autónomo ReAct de LangGraph inicializado con éxito y blindado contra errores de argumentos!")

except Exception as e:
    print(f"❌ Error al configurar el agente: {e}")
```

```python
import urllib.request
import xml.etree.ElementTree as ET
from langchain_core.tools import tool

@tool
def tool_cientifica(query: str) -> str:
    """Busca artículos científicos en arXiv y retorna sus títulos y resúmenes."""
    try:
        query_encoded = urllib.parse.quote(query)
        # Traemos solo 3 resultados para no saturar la pantalla de texto
        url = f"http://export.arxiv.org/api/query?search_query=all:{query_encoded}&max_results=3"

        with urllib.request.urlopen(url) as response:
            xml_data = response.read()

        root = ET.fromstring(xml_data)
        namespaces = {'atom': 'http://www.w3.org/2005/Atom'}
        entry_list = root.findall('atom:entry', namespaces)

        if not entry_list:
            return "No se encontraron artículos científicos."

        resultados = []
        for index, entry in enumerate(entry_list, 1):
            title = entry.find('atom:title', namespaces).text.strip().replace('\n', ' ')
            # EXTRAEMOS EL RESUMEN (Abstract)
            summary = entry.find('atom:summary', namespaces).text.strip().replace('\n', ' ')

            # Guardamos ambos datos formateados
            resultados.append(f"{index}. TÍTULO: {title}\n   RESUMEN: {summary[:300]}...\n") # Cortado a 300 caracteres para limpieza

        return "\n".join(resultados)

    except Exception as e:
        return f"Error al consultar la API de arXiv: {e}"

print("📚 Herramienta de arXiv actualizada (Títulos + Resúmenes).")
```

```python
from typing import Dict, TypedDict
from langgraph.graph import START, StateGraph, END

# 1. Definimos el Estado que compartirán los agentes
class EstructuraEstado(TypedDict):
    consulta_usuario: str
    papers_crudos: str
    reporte_final: str

# 2. Nodo 1: El Agente Investigador (Usa la herramienta de arXiv)
def agente_investigador(state: EstructuraEstado) -> Dict:
    print("🔬 [Investigador] Buscando papers relevantes en arXiv...")

    # Invocamos tu herramienta nativa pasándole la consulta
    query = state["consulta_usuario"]
    data_arxiv = tool_cientifica.invoke(query)

    # Guardamos el resultado en el estado
    return {"papers_crudos": data_arxiv}

# 3. Nodo 2: El Agente Redactor (Toma los datos y redacta el informe final)
def agente_redactor(state: EstructuraEstado) -> Dict:
    print("✍️ [Redactor] Procesando datos y armando el informe definitivo...")

    prompt_redaccion = f"""
    Actúa como un Redactor Científico Senior.
    Toma los siguientes artículos científicos crudos y escribe un reporte unificado, ordenado, profesional y fácil de leer para el usuario.

    PAPERS ENCONTRADOS:
    {state["papers_crudos"]}
    """

    # Usamos tu objeto model (Gemini) de forma directa
    respuesta = model.invoke(prompt_redaccion)

    # Guardamos la respuesta final en el estado
    return {"reporte_final": respuesta.content}

print("⚙️ Nodos y Estado configurados con éxito.")
```

```python
# 1. Inicializamos el grafo con nuestra estructura de estado
workflow = StateGraph(EstructuraEstado)

# 2. Añadimos los nodos al tablero
workflow.add_node("investigador_node", agente_investigador)
workflow.add_node("redactor_node", agente_redactor)

# 3. Creamos las conexiones (Aristas / Edges)
workflow.add_edge(START, "investigador_node")         # De START va directo al investigador
workflow.add_edge("investigador_node", "redactor_node") # Del investigador pasa al redactor
workflow.add_edge("redactor_node", END)                 # Del redactor termina el flujo (END)

# 4. Compilamos el flujo personalizado
app_multi_agente = workflow.compile()

print("🚀 ¡Grafo personalizado de LangGraph compilado y listo para producción!")
```

```python
# Definimos la entrada inicial con la consulta
estado_inicial = {
    "consulta_usuario": "neural networks applied to medicine",
    "papers_crudos": "",
    "reporte_final": ""
}

print("👤 Iniciando flujo de trabajo coordinado...")
print("-" * 50)

# Ejecutamos el grafo completo
resultado_flujo = app_multi_agente.invoke(estado_inicial)

print("-" * 50)
print("🤖 RESPUESTA COMPLETA Y COMBINADA:")
print("-" * 50)
print(resultado_flujo["reporte_final"])
```

```python
from typing import Literal

def ruteador_inteligente(state: EstructuraEstado) -> Literal["ir_a_investigar", "ir_a_charla"]:
    print("💻 [Ruteador Optimizado] Analizando consulta mediante palabras clave (0% API Cost)...")

    # Pasamos la consulta a minúsculas para evitar problemas de mayúsculas
    consulta = state["consulta_usuario"].lower()

    # Palabras clave que delatan una intención de búsqueda científica
    palabras_cientificas = ["paper", "papers", "arxiv", "artículo", "articulos", "investigación", "buscar", "busca"]

    # Evaluamos de forma determinista en Python puro
    if any(palabra in consulta for palabra in palabras_cientificas):
        print("➡️ Dirección: Flujo Científico (arXiv)")
        return "ir_a_investigar"
    else:
        print("➡️ Dirección: Responder directamente (Conversacional)")
        return "ir_a_charla"
```

```python
def agente_casual(state: EstructuraEstado) -> Dict:
    print("💬 [Asistente Casual] Respondiendo consulta general...")
    respuesta = model.invoke(state["consulta_usuario"])
    return {"reporte_final": respuesta.content}
```

```python
# 1. Inicializamos y agregamos los 3 nodos existentes
workflow_inteligente = StateGraph(EstructuraEstado)
workflow_inteligente.add_node("investigador_node", agente_investigador)
workflow_inteligente.add_node("redactor_node", agente_redactor)
workflow_inteligente.add_node("casual_node", agente_casual)

# 2. CONFIGURAMOS EL RUTEO CONDICIONAL DESDE EL START
workflow_inteligente.add_conditional_edges(
    START,                   # ¿De dónde sale el flujo? De START
    ruteador_inteligente,     # ¿Quién evalúa? Nuestra función ruteadora
    {                        # Mapeo: dependiendo de lo que retorne la función, a qué nodo va
        "ir_a_investigar": "investigador_node",
        "ir_a_charla": "casual_node"
    }
)

# 3. El resto de las conexiones fijas
workflow_inteligente.add_edge("investigador_node", "redactor_node")
workflow_inteligente.add_edge("redactor_node", END)
workflow_inteligente.add_edge("casual_node", END)

# 4. Compilamos
app_con_ruteador = workflow_inteligente.compile()
print("🚀 Grafo con Agente Ruteador compilado con éxito.")
```

```python
import gradio as gr

def run_graph(user_query):
    """
    Ejecuta el LangGraph personalizado con la pregunta del usuario
    y devuelve la respuesta final combinada.
    """
    estado_inicial = {
        "consulta_usuario": user_query,
        "papers_crudos": "",
        "reporte_final": ""
    }

    try:
        # Invocamos tu aplicación multi-agente
        result = app_con_ruteador.invoke(estado_inicial)
        return result["reporte_final"]

    except Exception as e:
        return f"❌ Ocurrió un error en el backend: {e}"

# Construimos la interfaz compatible con Gradio moderno
iface = gr.Interface(
    fn=run_graph,
    inputs=gr.Textbox(
        lines=3,
        placeholder="Ej: ¿Qué papers hay sobre transformers aplicados a la medicina?",
        label="Escribe tu pregunta:"
    ),
    outputs=gr.Markdown(label="Respuesta del Agente Científico"),
    title="📚 Agente de Búsqueda Científica Inteligente",
    description="Haz una consulta teórica o científica. El sistema decidirá mediante código si requiere indexar papers en arXiv o responder directamente.",

    # TRADUCCIÓN DE BOTONES AL CASTELLANO 🇦🇷
    submit_btn="Enviar Consulta",
    clear_btn="Limpiar",

    # FIX DE LA VERSIÓN NUEVA: Reemplazamos flagging_btn por flagging_mode
    # "manual" deja el botón por defecto (que se puede traducir internamente si usas bloques),
    # o podés poner "never" para ocultarlo por completo y limpiar la interfaz.
    flagging_mode="never"
)

# Lanzamos la interfaz
iface.launch(share=True)
```
